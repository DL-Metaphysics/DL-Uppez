### 批量梯度下降

```python
#引库
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

#将生成图片嵌入jupyter
%matplotlib inline

#忽略无意义的警告
import warnings
warnings.filterwarnings('ignore')

#用来正常显示中文标签
plt.rcParams['font.sans-serif']='SimHei'

#用来正常显示负号
plt.rcParams['axes.unicode_minus']=False

#读取训练集信息
dataone = pd.read_csv('dataone.csv')
x1 = np.array(dataone['area'])
y1 = np.array(dataone['price'])

#训练集图输出
plt.scatter(x1,y1)
plt.xlabel("x1", fontsize=18)
plt.ylabel("y1", fontsize=18 ,rotation=0)
plt.show()
```

```python
#θ的参数设置，随机取的
theta0=0
theta1=0

epsilon = 0.1

#alpha是学习率(步长),随机设定
alpha = 0.00001

#计算长度
m = len(x1)

result1 = []
count1 = 0

def h1(i):
    return theta0+theta1*x1[i]

def diff(i):
    return h1(i) - y1[i]

#f,ax=plt.subplots(1,2,figsize=(18,8))
while True:
    sum1 = 0
    sum2 = 0
    for i in range(m):
        sum1 = sum1 + diff(i)
        sum2 = sum2 + diff(i)*x1[i]
    theta0 = theta0 - (alpha/m)*sum1
    theta1 = theta1 - (alpha/m)*sum2
    
    loss = 0
    
    for i in range(m):
        loss += 0.5/m*(theta0+theta1*x1[i]-y1[i])**2
    count1 += 1
    result1.append(loss)
    print("theta0:{},theta1:{},loss:{},count:{}".format(theta0,theta1,loss,count1))
    
    if (loss<epsilon):
        break
    elif count1 == 1000:
        break
```

```python
for i in range(count1):
    plt.scatter(i,result1[i])
plt.xlabel("Generation")
plt.ylabel("Loss")
plt.show()
```

### 随机梯度下降

```python
#θ的参数设置，随机取的
theta0=1
theta1=2

#alpha是学习率(步长),随机设定
alpha = 0.00000001

#迭代次数
times = 10

#计算长度
m = len(x1)

def h1(i):
    return theta0+theta1*x1[i]

def diff(i):
    return h1(i) - y1[i]

for time in range(times):
    theta0 = theta0 - alpha*diff(i)*x1[i]
    theta1 = theta1 - alpha*diff(i)*x1[i]
    
print("theta0:",theta0)
print("theta1:",theta1)
```
