### 批量梯度下降

```python
#引库
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

#将生成图片嵌入jupyter
%matplotlib inline

#忽略无意义的警告
import warnings
warnings.filterwarnings('ignore')

#用来正常显示中文标签
plt.rcParams['font.sans-serif']='SimHei'

#用来正常显示负号
plt.rcParams['axes.unicode_minus']=False

#读取训练集信息
dataone = pd.read_csv('dataone.csv')
x1 = np.array(dataone['area'])
y1 = np.array(dataone['price'])

#训练集图输出
plt.scatter(x1,y1)
plt.xlabel("x1", fontsize=18)
plt.ylabel("y1", fontsize=18 ,rotation=0)
plt.show()
```

```python
#θ的参数设置，随机取的
theta0=1
theta1=2

#alpha是学习率(步长),随机设定
alpha = 0.00000001

#迭代次数
times = 1000

#计算长度
m = len(x1)

def h1(i):
    return theta0+theta1*x1[i]

def diff(i):
    return h1(i) - y1[i]

#f,ax=plt.subplots(1,2,figsize=(18,8))
for time in range(times):
    sum1 = 0
    sum2 = 0
    for i in range(m):
        sum1 = sum1 + diff(i)
        sum2 = sum2 + diff(i)*x1[i]
    theta0 = theta0 - (alpha/m)*sum1
    theta1 = theta1 - (alpha/m)*sum2
    #作图
    '''plt.subplot(121)
    plt.scatter(i,theta0)
    plt.subplot(122)
    plt.scatter(i,theta1)'''
    
print("theta0:",theta0)
print("theta1:",theta1)
#plt.show()
```
